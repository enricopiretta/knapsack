# ALGORITHM

```python
# Enter your code for the Greedy Cow Transport here 
def greedy_cow_transport(cows,limit=10):
    value_list = []
    key_list = []
    for b in cows.values():
        value_list.append(b)
    for b in cows.keys():
        key_list.append(b)
    draft_list = []
    for u in range(len(value_list)):
        draft_list.append([value_list[u],key_list[u]])
    e = sorted(draft_list, reverse = True)
    final_trip = []
    while not(not e):
        trip = []
        e_copy = e[:]
        tot_cost = 0.0
        for i in e_copy:
            c = []
            if (tot_cost + i[0]) <= limit:
                c.append(i)
                trip.append(i[1])
                tot_cost += i[0]
                e.remove(c[0])
        final_trip.append(trip)
    return final_trip
```
